```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import scipy.ndimage as im

%matplotlib inline
sns.set_style('whitegrid')
# sns.set(font='Source Sans Pro Light')
```

```python
%ls ../results
```

```python
aug = pd.read_csv('../results/run_augmented1,tag_val_acc.csv')
ref = pd.read_csv('../results/run_reference1,tag_val_acc.csv')
aug.head()
```

```python
xaug = np.arange(len(aug['Value']))
xaug = xaug * 309

xref = np.arange(len(ref['Value']))
xref = xref * 53*3

plt.figure(figsize=(9,6))
plt.plot(xaug, aug['Value'])
plt.plot(xref, ref['Value'])
plt.gca().set_ylim([0,1])
plt.title('Validation Accuracy Comparison')
plt.legend(['Augmented Dataset', 'Original Dataset'])
plt.ylabel('Validation Accuracy')
plt.xlabel('Samples Trained')
```

```python
print 'ref:',ref['Value'].max(), 'aug:', aug['Value'].max()
```

```python
aug = pd.read_csv('../results/run_augmented1,tag_val_loss.csv')
ref = pd.read_csv('../results/run_reference1,tag_val_loss.csv')

xaug = np.arange(len(aug['Value']))
xaug = xaug * 309

xref = np.arange(len(ref['Value']))
xref = xref * 53*3

plt.figure(figsize=(9,6))
plt.plot(xaug, aug['Value'])
plt.plot(xref, ref['Value'])
plt.gca().set_ylim([0,2])
plt.title('Validation Loss Comparison')
plt.legend(['Augmented Dataset', 'Original Dataset'])
plt.ylabel('Validation Loss')
plt.xlabel('Samples Trained')
```
